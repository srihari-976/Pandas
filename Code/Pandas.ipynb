```python
vk = deli[deli['batsman'] == 'V Kohli']
```

```python
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head()
```

```python
# function to predict the highest runs made by the players
def Highruns(batsman):
    vk = deli[deli['batsman'] == batsman]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
    
```

```python
Highruns('MS Dhoni')
```

```python
# find the most destructive batsman in the history of the IPL
# Strike Rate = (Number of runs/Number of balls)/100
# Min batsman 200 balls in over 16-20
```

```python
destru=deli[deli['over']>15]
```

```python
deli[deli['over']>15].shape
```

```python
destru.groupby('batsman')['batsman_runs'].count()
```

```python
all_batsman=destru.groupby('batsman')['batsman_runs'].count()
x = all_batsman>200
batsman_list=all_batsman[x].index.tolist()
```

```python
final=deli[deli['batsman'].isin(batsman_list)]
```

```python
deli[deli['batsman'].isin(batsman_list)].shape
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
```
